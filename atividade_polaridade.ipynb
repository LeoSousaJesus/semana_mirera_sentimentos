## Principais funcionalidades do TextBlob:

1. Análise de Sentimentos: Permite determinar se um texto tem uma conotação positiva, negativa ou neutra, além de fornecer uma pontuação de subjetividade.

2. Tradução e Detecção de Idioma: Facilita a tradução de texto entre diferentes idiomas e a detecção do idioma de um texto.

3. Extração de Frases e Nomes: Possui métodos para identificar e extrair frases, palavras-chave, e entidades nomeadas.

4. Correção Gramatical: Pode corrigir erros de digitação ou gramática em um texto.

5. Tokenização: Divide o texto em palavras e frases, facilitando a análise.

```python
from textblob import TextBlob
import spacy
from spacy import displacy

# Criando um objeto TextBlob com o texto em português
texto = TextBlob("Eu adoro programar em Python!")

# Análise de sentimentos (polaridade e subjetividade)
analysis = texto.sentiment  # O método 'sentiment' já retorna a polaridade e subjetividade
print("Polaridade e Subjetividade:", analysis)

# Tradução do texto do português para o inglês
texto_traduzido = texto.translate(from_lang='pt', to='en')
print("Texto traduzido:", texto_traduzido)

# Extraindo frases e palavras-chave
frases = texto.sentences
palavras = texto.words
print("Frases:", frases)
print("Palavras:", palavras)

# Exemplo de polaridade e subjetividade com base na análise do TextBlob
polaridade = analysis.polarity  # Obtém a polaridade do texto
subjetividade = analysis.subjectivity  # Obtém a subjetividade do texto

print("Polaridade:", polaridade)
print("Subjetividade:", subjetividade)
```

# No TextBlob, a polaridade é uma medida usada na análise de sentimentos que indica a conotação emocional de um texto. A polaridade varia de -1 a 1:

## Interpretação


#### -1: Sentimento muito negativo
- Valores próximos de -1 indicam que o texto é muito negativo.<br><br>

#### 0: Sentimento neutro
- Valores em torno de 0 sugerem neutralidade.<br><br>

#### 1: Sentimento muito positivo
- Valores próximos de 1 indicam que o texto é muito positivo.








## Como funciona a polaridade
- O TextBlob atribui uma pontuação de polaridade a um texto com base nas palavras e nas suas relações contextuais. Palavras positivas, como "bom" ou "feliz", contribuem para uma pontuação positiva, enquanto palavras negativas, como "ruim" ou "triste", resultam em uma pontuação negativa.

### Considerações
- Contexto: A polaridade pode ser influenciada pelo contexto e pela combinação de palavras, então frases complexas podem gerar resultados que não refletem exatamente a intenção do autor.
- Limitações: TextBlob é eficaz para textos simples, mas pode ter dificuldades com sarcasmo ou ironia.

```python
# Frases de exemplo e suas polaridades verdadeiras
dados = [
    ("O filme foi incrível!", 1),   # Positivo
    ("Eu odiei essa comida.", -1),   # Negativo
    ("O dia está ok.", 0),           # Neutro
    ("Que ótimo trabalho!", 1),       # Positivo
    ("Foi um péssimo atendimento.", -1) # Negativo
]

# Contadores para calcular a acurácia
total = len(dados)
corretos = 0

from textblob import TextBlob

# Avaliando a acurácia
for frase, polaridade_esperada in dados:
    texto = TextBlob(frase)
    polaridade_calculada = texto.sentiment.polarity

    # Definindo a polaridade predita com base no valor calculado
    if polaridade_calculada > 0:
        polaridade_predita = 1
    elif polaridade_calculada < 0:
        polaridade_predita = -1
    else:
        polaridade_predita = 0

    # Comparando a polaridade predita com a polaridade esperada
    if polaridade_predita == polaridade_esperada:
        corretos += 1

    # Imprimindo a polaridade calculada e a polaridade esperada para cada frase
    print(f"Frase: {frase}")
    print(f"Polaridade esperada: {polaridade_esperada}, Polaridade calculada: {polaridade_calculada:.2f}, Polaridade predita: {polaridade_predita}")
    print()

# Calculando a acurácia
acuracia = corretos / total
print(f'Acurácia: {acuracia:.2f}')  # Exemplo de saída: Acurácia: 1.00
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Carregar o DataFrame
df = pd.read_csv("dados/industrializacao.csv")

# Função para calcular a polaridade
def calcular_polaridade(comentario):
    return TextBlob(comentario).sentiment.polarity

# Adicionar uma nova coluna com a polaridade
df['polaridade'] = df['comments'].apply(calcular_polaridade)

# Exibir os comentários e suas polaridades
print(df[['comments', 'polaridade']])

# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(df['comments'], df['polaridade'], color='skyblue')
plt.xlabel('Polaridade')
plt.title('Análise de Sentimentos dos Comentários sobre Industrialização')
plt.axvline(0, color='red', linestyle='--')  # Linha para a polaridade zero
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrar o gráfico
plt.show()
```
